## Task Overview

Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, such as the disctribution of ages or genders in a population.

```python
# Install required libraries

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the dataset from the source

data = pd.read_csv("/content/PRODIGY_DS_01/WorldPopulation.csv")
```

```python
# Initial display of dataset

data.head(10)
```

```python
print(data.describe)
```

```python
#List of contries for the graph from all continentس
countries_to_plot = ['Italy','United States', 'Canada', 'Brazil', 'Japan', 'Germany', 'France', 'Iran, Islamic Rep.', 'South Africa', 'Australia']

#Filtering the data frame
filtered_data = data[data['Country Name'].isin(countries_to_plot)]

countries_to_plot
```

```python
# Set a larger figure size for better readability
plt.figure(figsize=(10, 6))

# Creating the bar chart with a specific color and edge color for bars
plt.bar(filtered_data['Country Name'], filtered_data['2022'], color='skyblue', edgecolor='black', label='Population')

# Adding a line plot (distribution line) to show the trend
plt.plot(filtered_data['Country Name'], filtered_data['2022'], color='red', marker='o', linewidth=2, label='Trend Line')

# Adding labels and title with improved font size and styling
plt.xlabel('Country Name', fontsize=12, fontweight='bold')
plt.ylabel('Population in 2022 (in billions)', fontsize=12, fontweight='bold')
plt.title('Population of Selected Countries in 2022', fontsize=14, fontweight='bold')

# Adding gridlines for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Rotating the x-axis labels for better readability
plt.xticks(rotation=45, ha='right', fontsize=10)

# Adding a legend to distinguish between bar chart and line plot
plt.legend()

# Adding tight layout to avoid overlapping of elements
plt.tight_layout()

# Displaying the plot
plt.show()
```

This chart compares the populations of selected countries in 2022 using blue bars to represent population size and a red line to show the trend across these countries. The United States has the largest population, approaching 300 million, followed by Brazil with about 200 million. Other countries like Germany, Japan, and Iran have populations around 80-100 million, while countries like Australia, Canada, and South Africa have much smaller populations, around 50 million or less. The red trend line highlights the fluctuations in population across the countries, rising and falling depending on the country’s population size.
